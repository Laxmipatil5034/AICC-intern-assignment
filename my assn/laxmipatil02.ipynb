```python
# ML Idea Generator

def generate_ml_ideas():
    ideas = [
        {
            "domain": "College",
            "problem": "Student Performance Prediction",
            "input": "Attendance, internal marks, study hours, assignment scores",
            "output": "Predicted final exam score or pass/fail"
        },
        {
            "domain": "Healthcare",
            "problem": "Disease Prediction System",
            "input": "Symptoms, age, medical history, test reports",
            "output": "Predicted disease or health risk"
        },
        {
            "domain": "Shopping",
            "problem": "Product Recommendation System",
            "input": "User purchase history, search history, ratings",
            "output": "Recommended products"
        },
        {
            "domain": "College",
            "problem": "Attendance Monitoring System",
            "input": "Face images or RFID logs",
            "output": "Attendance status (Present/Absent)"
        },
        {
            "domain": "Healthcare",
            "problem": "Heart Disease Prediction",
            "input": "Blood pressure, cholesterol, ECG results, age",
            "output": "Risk level (Low/Medium/High)"
        }
    ]

    return ideas


def display_ideas(ideas):
    print("---- ML Idea Generator Output ----\n")
    for i, idea in enumerate(ideas, 1):
        print(f"Idea {i}:")
        print(f"Domain  : {idea['domain']}")
        print(f"Problem : {idea['problem']}")
        print(f"Input   : {idea['input']}")
        print(f"Output  : {idea['output']}")
        print("-" * 40)


# Main execution
ml_ideas = generate_ml_ideas()
display_ideas(ml_ideas)
```
