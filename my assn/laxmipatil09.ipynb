```python
# Import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression

# Step 1: Create Dataset
data = {
    "Area_sqft": [500, 700, 900, 1100, 1300, 1500],
    "Price": [1000000, 1500000, 1800000, 2200000, 2600000, 3000000]
}

df = pd.DataFrame(data)

print("Dataset:\n")
print(df)

# Step 2: Features and Labels
X = df[["Area_sqft"]]   # Feature
y = df["Price"]         # Label

# Step 3: Train Model
model = LinearRegression()
model.fit(X, y)

# Step 4: Predict new house price
new_area = [[1200]]
predicted_price = model.predict(new_area)

print("\nPrediction:")
print("Price for 1200 sqft house =", int(predicted_price[0]))
```
