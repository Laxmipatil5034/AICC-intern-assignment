```python
# Assignment: Student Data Manager
# Date: 19/02/2026

# 1. Store data for 5 students using a list of dictionaries
students = [
    {"name": "Amit", "score": 85},
    {"name": "Sana", "score": 92},
    {"name": "John", "score": 78},
    {"name": "Priya", "score": 95},
    {"name": "Rahul", "score": 64}
]

def manage_student_data(student_list):
    total_score = 0
    topper = student_list[0] # Assume the first one is the topper initially

    print(f"{'Name':<10} | {'Score':<6} | {'Grade':<5}")
    print("-" * 30)

    for student in student_list:
        score = student['score']
        total_score += score
        
        # Determine Grade
        if score >= 90:
            grade = 'A'
        elif score >= 80:
            grade = 'B'
        elif score >= 70:
            grade = 'C'
        elif score >= 60:
            grade = 'D'
        else:
            grade = 'F'
            
        print(f"{student['name']:<10} | {score:<6} | {grade:<5}")

        # Check for topper
        if score > topper['score']:
            topper = student

    # 2. Calculate Class Average
    average = total_score / len(student_list)

    print("-" * 30)
    print(f"Class Average: {average:.2f}")
    print(f"Class Topper:  {topper['name']} with {topper['score']}%")

# Call the function
manage_student_data(students)
```
