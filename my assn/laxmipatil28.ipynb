```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Setting the aesthetic style for the plots
sns.set_theme(style="whitegrid")

# --- 1. Create Sample Data ---
data = {
    'Category': ['Rent', 'Food', 'Transport', 'Entertainment', 'Savings', 'Utilities'],
    'Amount': [800, 450, 150, 200, 300, 100]
}
df = pd.DataFrame(data)

# Generate 100 random transaction values for the histogram (daily spending habits)
np.random.seed(42)
daily_transactions = np.random.gamma(shape=2, scale=20, size=100) 

# --- 2. Create the Visualizations ---
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# A. Bar Chart: Comparing Categories
sns.barplot(x='Category', y='Amount', data=df, ax=axes[0], palette='viridis')
axes[0].set_title('Monthly Expenses by Category')
axes[0].set_ylabel('Amount ($)')

# B. Pie Chart: Proportions of Total Budget
axes[1].pie(df['Amount'], labels=df['Category'], autopct='%1.1f%%', 
            startangle=140, colors=sns.color_palette('pastel'))
axes[1].set_title('Budget Distribution (%)')

# C. Histogram: Frequency of Transaction Sizes
sns.histplot(daily_transactions, bins=15, kde=True, ax=axes[2], color='skyblue')
axes[2].set_title('Distribution of Daily Transaction Sizes')
axes[2].set_xlabel('Transaction Amount ($)')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
