```python
def fizz_buzz():
    fizz_count = 0
    buzz_count = 0
    fizzbuzz_count = 0
    number_count = 0

    for i in range(1, 51):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz", end=" ")
            fizzbuzz_count += 1
        elif i % 3 == 0:
            print("Fizz", end=" ")
            fizz_count += 1
        elif i % 5 == 0:
            print("Buzz", end=" ")
            buzz_count += 1
        else:
            print(i, end=" ")
            number_count += 1

    # Print counts
    print("\n\n--- Counts ---")
    print("Fizz count:", fizz_count)
    print("Buzz count:", buzz_count)
    print("FizzBuzz count:", fizzbuzz_count)
    print("Numbers count:", number_count)

# THIS IS THE MISSING PIECE:
# Calling the function to execute the code above
fizz_buzz()
```
