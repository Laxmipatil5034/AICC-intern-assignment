```python
# Import libraries
import pandas as pd
from sklearn.neighbors import NearestNeighbors

# Step 1: Create dataset (User preferences)
# 1 = likes, 0 = not interested

data = {
    "Action":   [1, 1, 0, 0, 1],
    "Comedy":   [1, 0, 1, 0, 1],
    "Romance":  [0, 0, 1, 1, 0],
    "Horror":   [0, 1, 0, 1, 0]
}

users = ["User1", "User2", "User3", "User4", "User5"]

df = pd.DataFrame(data, index=users)

print("User Preference Dataset:\n")
print(df)

# Step 2: Train KNN model
model = NearestNeighbors(n_neighbors=2, metric='euclidean')
model.fit(df)

# Step 3: Find similar users for a new user
# New user likes Action & Comedy
new_user = [[1, 1, 0, 0]]

distances, indices = model.kneighbors(new_user)

print("\nNearest Users:")
for i in indices[0]:
    print(users[i])
```
