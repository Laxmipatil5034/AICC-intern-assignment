```python
import hashlib

# Function to hash password
def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()

# Predefined username and password (stored securely as hash)
stored_username = "admin"
stored_password_hash = hash_password("Secure@123")

# Maximum login attempts
max_attempts = 3
attempts = 0

print("=== Login System ===")

while attempts < max_attempts:
    username = input("Enter Username: ")
    password = input("Enter Password: ")

    # Hash entered password
    entered_password_hash = hash_password(password)

    # Check credentials
    if username == stored_username and entered_password_hash == stored_password_hash:
        print("✅ Login Successful!")
        break
    else:
        attempts += 1
        print(f"❌ Incorrect credentials! Attempts left: {max_attempts - attempts}")

# If all attempts are used
if attempts == max_attempts:
    print("🔒 Account locked due to too many failed attempts.")
```
