```python
import numpy as np
import time

# Assignment: NumPy Speed Test
# Date: 20/02/2026

# 1. Initialize data with 1 million numbers
size = 1_000_000
python_list = list(range(size))
numpy_array = np.array(python_list)

print(f"Comparing performance for {size:,} elements...\n")

# --- Python List Timing ---
start_time = time.time()
# Standard approach: List comprehension to add 1 to each element
list_result = [x + 1 for x in python_list]
python_duration = time.time() - start_time

print(f"Python List Time: {python_duration:.5f} seconds")

# --- NumPy Array Timing ---
start_time = time.time()
# NumPy approach: Vectorized addition
array_result = numpy_array + 1
numpy_duration = time.time() - start_time

print(f"NumPy Array Time: {numpy_duration:.5f} seconds")

# 2. Performance Comparison
speed_factor = python_duration / numpy_duration
print(f"\nNumPy is approximately {speed_factor:.1f}x faster than a Python list.")
```
