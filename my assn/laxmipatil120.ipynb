```python
from sklearn import tree
import matplotlib.pyplot as plt

# Sample dataset
# Weather: 0=Sunny, 1=Rainy, 2=Cloudy
# Humidity: 0=High, 1=Normal
# Wind: 0=No, 1=Yes

X = [
    [0, 0, 0],  # Sunny, High, No
    [0, 1, 0],  # Sunny, Normal, No
    [1, 0, 1],  # Rainy, High, Yes
    [1, 1, 0],  # Rainy, Normal, No
    [2, 1, 0]   # Cloudy, Normal, No
]

y = [0, 1, 0, 1, 1]  # 0=Don't Play, 1=Play

model = tree.DecisionTreeClassifier()
model.fit(X, y)

# Plot tree
plt.figure()
tree.plot_tree(model, feature_names=["Weather","Humidity","Wind"],
               class_names=["No","Yes"],
               filled=True)

plt.show()
```
