```python
# Import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression

# Step 1: Create Dataset
data = {
    "Study_Hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "Marks": [35, 40, 50, 55, 65, 70, 75, 85]
}

df = pd.DataFrame(data)

# Display dataset
print("Dataset:\n")
print(df)

# Step 2: Identify Features (X) and Labels (y)
X = df[["Study_Hours"]]   # Feature
y = df["Marks"]           # Label

# Step 3: Train Model
model = LinearRegression()
model.fit(X, y)

# Step 4: Predict relationship
predicted = model.predict([[9]])

print("\nPrediction:")
print("If a student studies 9 hours, predicted marks =", round(predicted[0], 2))
```
